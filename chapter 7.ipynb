```python
from scipy import interpolate
from numpy import polynomial as P
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
p1 = P.Polynomial([1,2,3])     #Polynomial代表类即class
p1                             #p1是Polynomial的对象即object
```

```python
np.__version__         #查看numpy的版本
```

```python
p2 = P.Polynomial.fromroots([-1, 1])
p2
```

```python
p1.roots()
```

```python
p2.roots()
```

```python
x = np.linspace(-5,5)     
plt.plot(x,p1(x))
plt.plot(x,p2(x))              #前三行是画函数p1,p2图像的命令
plt.ylim([-5,20])              #设置纵坐标y的区间
```

```python
fig=plt.figure(figsize=(10,8))
x=np.linspace(-5,5)
plt.plot(x,p1(x),label=str(p1))
plt.plot(x,p2(x),label=str(p2))
plt.axhline(0,ls=':')
plt.legend()
ax_insert=fig.add_axes([0.35,0.4,0.35,0.45])
ax_insert.plot(x,p1(x))
ax_insert.plot(x,p2(x))
ax_insert.set_xlim([-2,2])
ax_insert.set_ylim([-2,6])
ax_insert.axhline(0,ls=':')
```
